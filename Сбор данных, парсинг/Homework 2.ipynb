```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://books.toscrape.com/'
req = requests.get(url)
req.status_code
```

```python
soup = bs(req.content, 'html.parser')
books = soup.select('article')
```

```python
#Формируется список названий

titles = []

for book in books:
    h3 = book.find('h3')
    title = h3.find('a')['title']
    titles.append(title)
```

```python
#Формируется список цен

prices = []

for price in books:
    div = price.find('div', attrs={'class':'product_price'})
    number = div.find('p', attrs={'class':'price_color'}).text
    prices.append(float(number[1:]))
```

```python
#Формируется список рейтингов

ratings = []
words = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}

for rating in books:
    p = rating.find('p')
    ratings.append(p['class'][1])
    
ratings = list(map(words.get, ratings))
```

```python
#Создание датасета

shop = pd.DataFrame({
    'Title': titles,
    'Rating': ratings,
    'Price': prices
})

shop
```
